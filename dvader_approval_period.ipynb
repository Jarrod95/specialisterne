```python
import pandas as pd
import datetime
```

```python
cyberark = pd.read_excel(r'C:\Users\Jarrod\Documents\Data Science\Specialistinterne Workshop\SIEM Task\wbc_siem_cyberark_dataset.xlsx')

#Seperate instances of 'dvader' as a destination user name
dvader = cyberark[cyberark['destinationUserName'].str.contains('dvader')]

#Drop 'AEST' from deviceCustomDate columns
dvader['Event Time'] = dvader['Event Time'].str.replace('AEST', '')
dvader['deviceCustomDate1'] = dvader['deviceCustomDate1'].str.replace('AEST', '')
dvader['deviceCustomDate2'] = dvader['deviceCustomDate2'].str.replace('AEST', '')

#Ensure datetime columns are read as datetime 
dvader['Event Time'] = pd.to_datetime(dvader['Event Time'], format='%Y-%m-%d %H:%M:%S')
dvader['deviceCustomDate1'] = pd.to_datetime(dvader['deviceCustomDate1'], format='%Y-%m-%d %H:%M:%S')
dvader['deviceCustomDate2'] = pd.to_datetime(dvader['deviceCustomDate2'], format='%Y-%m-%d %H:%M:%S')

#Sort DataFrame
dvader.sort_values('deviceCustomDate1', ascending=True, inplace=True)
dvader.reset_index(drop=True)
```

```python
#Calculate time difference between 'log in' and 'log off'
dvader['time_diff'] = dvader['deviceCustomDate2'] - dvader['deviceCustomDate1']

#find maximum in 'time_diff' column
print(dvader['time_diff'].max())
```
